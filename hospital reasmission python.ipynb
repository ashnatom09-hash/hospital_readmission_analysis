```python


categorical_cols = ['age','medical_specialty','glucose_test','A1Ctest','change','diabetes_med']

for col in categorical_cols:
    print("Column:", col)
    print(pd.crosstab(df[col], df['readmitted'], normalize='index')*100)
    print("\n"+"-"*50+"\n")
```

```python
# 1️⃣ Define numeric columns
numeric_cols = [
    'time_in_hospital',
    'n_lab_procedures',
    'n_procedures',
    'n_medications',
    'n_outpatient',
    'n_inpatient',
    'n_emergency'
]

# 2️⃣ Loop through each numeric column and calculate average per readmitted group
for col in numeric_cols:
    print(f"Column: {col}")
    print(df.groupby('readmitted')[col].mean())
    print("-"*50)
```
